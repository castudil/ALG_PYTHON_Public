* Complejidad teórica: Tiempo $\mathcal{O}(n)$, Espacio $\mathcal{O}(1)$.

```python
def busqueda_maximo(arreglo):
    if not arreglo:
        return None
    
    maximo = arreglo[0]
    for elemento in arreglo:
        if elemento > maximo:
            maximo = elemento
            
    return maximo

temperaturas = [15.5, 18.2, 14.1, 22.0, 19.3]
print(f"Temperatura máxima detectada: {busqueda_maximo(temperaturas)}")
```

### Bloque 4: Verificación Empírica de Complejidad
Comprobación del modelo asintótico lineal $\mathcal{O}(n)$ mediante la inyección de arreglos de gran magnitud. Ejecuta la celda y analiza el output.

```python
import random
import time

def medir_tiempo_ejecucion(tamano_n):
    # Generación programática del arreglo en memoria
    datos_prueba = [random.uniform(0.0, 40.0) for _ in range(tamano_n)]
    
    inicio = time.perf_counter()
    _ = busqueda_maximo(datos_prueba)
    fin = time.perf_counter()
    
    tiempo_transcurrido = fin - inicio
    return tiempo_transcurrido

# Prueba asintótica
n_base = 1_000_000
n_escala = 10_000_000

print("Iniciando evaluación empírica...")
tiempo_base = medir_tiempo_ejecucion(n_base)
print(f"Tiempo para n={n_base:,}: {tiempo_base:.4f} segundos")

tiempo_escala = medir_tiempo_ejecucion(n_escala)
print(f"Tiempo para n={n_escala:,}: {tiempo_escala:.4f} segundos")

incremento_n = n_escala / n_base
incremento_tiempo = tiempo_escala / tiempo_base
print(f"\nAnálisis: Al aumentar la entrada en {incremento_n}x, el tiempo aumentó en {incremento_tiempo:.2f}x. Esto comprueba empíricamente O(n).")
```
