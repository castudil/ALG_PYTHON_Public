### 1. Evolución del Método Científico
El método científico es un proceso iterativo de observación, hipótesis, experimentación y revisión, desarrollado a través de cuatro etapas clave:
* **Antecedentes Antiguos:** Pensamiento empírico en Egipto (c. 1600 a.C.) y razonamiento deductivo de Aristóteles.
* **Edad de Oro del Islam:** Alhacén (1021 d.C.) y la necesidad de pruebas experimentales reproducibles combinadas con lógica matemática.
* **Renacimiento:** Francis Bacon (1620) y la formalización del método inductivo.
* **Revolución Científica:** Galileo, Descartes y Newton consolidan la matematización, el análisis racional y el modelo de leyes universales.

### 2. Taxonomía de Problemas Computacionales
Clasificación estructurada bajo la Teoría de la Computabilidad y Complejidad.

**Según la Naturaleza del Resultado:**
* **De Decisión:** Respuesta binaria (Sí/No).
* **De Búsqueda:** Localización de una estructura o valor.
* **De Optimización:** Búsqueda de la mejor solución (minimizar/maximizar).
* **De Conteo:** Cálculo de la cardinalidad de soluciones válidas.

**Según Computabilidad:**
* **Decidibles:** Existe un algoritmo que siempre se detiene.
* **Indecidibles:** No existe un algoritmo general (ej. Problema de la Parada).

**Según Complejidad Computacional:**
* **P:** Resolubles en tiempo polinómico determinista $\mathcal{O}(n^k)$.
* **NP:** Soluciones verificables en tiempo polinómico determinista.
* **NP-Completo:** Los problemas más difíciles de NP.
* **NP-Hard:** Al menos tan difíciles como los NP-Completos, no restringidos a decisión.

### 3. Problemas de Conteo (Clase #P)
Buscan determinar la cantidad exacta de soluciones válidas.
1. **Conteo de Caminos Simples:** Calcular el número total de rutas distintas entre dos nodos en un grafo sin repetir nodos.
2. **Conteo de Coloraciones:** Calcular las formas de asignar $k$ colores a los nodos de un grafo sin que nodos adyacentes compartan color.
3. **Conteo de Satisfacibilidad (#SAT):** Calcular las asignaciones de valores booleanos que hacen verdadera una fórmula lógica.

### 4. Importancia de Algoritmos y Estructuras de Datos

### 6. Contraste Práctico: Implementación en Python
Relación directa entre una estructura de datos (arreglo dinámico) y la ejecución de un algoritmo de búsqueda lineal.

```python
# Estructura de Datos: Lista (Arreglo dinámico en memoria)
temperaturas = [15.5, 18.2, 14.1, 22.0, 19.3]

# Algoritmo: Búsqueda lineal del valor máximo
max_temp = temperaturas[0]
for temp in temperaturas:
    if temp > max_temp:
        max_temp = temp

print(f"Temperatura máxima: {max_temp}")

# Análisis de Complejidad:
# Tiempo: O(n) - Itera sobre todos los elementos una vez.
# Espacio: O(1) - Utiliza una variable extra independiente del tamaño del arreglo.
```
