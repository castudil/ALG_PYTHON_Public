## 📖 Parte 1: ¿Cómo funciona el Ordenamiento por Burbuja?

El **ordenamiento por burbuja** (*Bubble Sort*) es uno de los algoritmos de ordenamiento más simples e intuitivos que existen. Su nombre proviene de la analogía con burbujas en el agua: los elementos más grandes "burbujean" hacia el final de la lista, pasaje a pasaje, de la misma manera en que las burbujas de aire ascienden a la superficie.

### Idea general

El algoritmo recorre la lista de elementos repetidamente. En cada recorrido, compara **pares de elementos adyacentes** y los **intercambia** si están en el orden incorrecto (es decir, si el elemento de la izquierda es mayor que el de la derecha). Este proceso se repite hasta que ningún intercambio sea necesario, lo que indica que la lista ya está ordenada.

### Descripción paso a paso

Supongamos que tenemos el siguiente arreglo de 5 elementos:

```
[ 5,  3,  8,  1,  4 ]
```

**Pasada 1:**
- Comparamos el par (5, 3) → 5 > 3, los intercambiamos → `[ 3, 5, 8, 1, 4 ]`
- Comparamos el par (5, 8) → 5 < 8, no se intercambian → `[ 3, 5, 8, 1, 4 ]`
- Comparamos el par (8, 1) → 8 > 1, los intercambiamos → `[ 3, 5, 1, 8, 4 ]`
- Comparamos el par (8, 4) → 8 > 4, los intercambiamos → `[ 3, 5, 1, 4, 8 ]`
- ✅ El elemento más grande (8) ya está en su posición final.

**Pasada 2:**
- Comparamos (3, 5) → no se intercambia
- Comparamos (5, 1) → intercambiamos → `[ 3, 1, 5, 4, 8 ]`
- Comparamos (5, 4) → intercambiamos → `[ 3, 1, 4, 5, 8 ]`
- ✅ El segundo elemento más grande (5) está en su posición final.

**Pasada 3:**
- Comparamos (3, 1) → intercambiamos → `[ 1, 3, 4, 5, 8 ]`
- Comparamos (3, 4) → no se intercambia
- ✅ El tercer elemento más grande (4) está en su posición final.

**Pasada 4:**
- Comparamos (1, 3) → no se intercambia
- ✅ La lista ya está ordenada: `[ 1, 3, 4, 5, 8 ]`

### Observaciones clave

- Después de cada pasada completa, el elemento más grande de la porción no ordenada queda en su posición correcta al final. Por eso, en cada pasada siguiente se necesita comparar un elemento menos.
- Para un arreglo de **n** elementos, el algoritmo realiza a lo sumo **n-1 pasadas**.
- En el **peor caso** (arreglo ordenado en orden inverso), se realizan aproximadamente $\frac{n(n-1)}{2}$ comparaciones, lo que da una complejidad de $\mathcal{O}(n^2)$.
- En el **mejor caso** (arreglo ya ordenado), una versión optimizada del algoritmo puede detectar que no hubo intercambios y terminar en una sola pasada, logrando $\mathcal{O}(n)$.

### Pseudocódigo

```
procedimiento burbuja(A, n):
    para i desde 0 hasta n-2:
        para j desde 0 hasta n-2-i:
            si A[j] > A[j+1]:
                intercambiar A[j] y A[j+1]
```

> 💡 **Nota:** En el pseudocódigo, el índice `i` representa la pasada actual. A medida que `i` aumenta, el bucle interno recorre un rango más corto porque los últimos `i` elementos ya están en su posición correcta.

````python
# ============================================================
# Función auxiliar: generar un arreglo aleatorio
# ============================================================
def generar_arreglo(n, valor_min=1, valor_max=100000):
    """
    Genera una lista de n números enteros aleatorios.
    
    Parámetros:
    -----------
    n         : tamaño del arreglo
    valor_min : valor mínimo posible (por defecto 1)
    valor_max : valor máximo posible (por defecto 100000)
    
    Retorna:
    --------
    Una lista de n enteros aleatorios.
    """
    return [random.randint(valor_min, valor_max) for _ in range(n)]

# ============================================================
# Función auxiliar: medir tiempo de ejecución
# ============================================================
def medir_tiempo(funcion_ordenamiento, arreglo):
    """
    Mide el tiempo de ejecución de una función de ordenamiento.
    
    Parámetros:
    -----------
    funcion_ordenamiento : función que recibe un arreglo y lo ordena
    arreglo              : lista de elementos a ordenar
    
    Retorna:
    --------
    tiempo_segundos : tiempo de ejecución en segundos (float)
    """
    # Hacemos una copia para no modificar el arreglo original
    copia = arreglo.copy()
    
    # Registramos el tiempo antes de ordenar
    inicio = time.time()
    
    # Ejecutamos el algoritmo
    funcion_ordenamiento(copia)
    
    # Registramos el tiempo después de ordenar
    fin = time.time()
    
    # Calculamos la diferencia
    tiempo_segundos = fin - inicio
    return tiempo_segundos


# ============================================================
# Definición de los tamaños de arreglo a probar
# ============================================================
tamanios = [1000, 2000, 4000, 8000, 16000]

print("✅ Funciones auxiliares definidas.")
print(f"   Tamaños a probar: {tamanios}")
````

## 📈 Parte 6: Análisis de Complejidad Experimental (Estrategia de Sedgewick)

### ¿Qué es la estrategia de Sedgewick?

Robert Sedgewick propone una manera simple y poderosa de determinar empíricamente el orden de crecimiento de un algoritmo. La idea es la siguiente:

Si el tiempo de ejecución de un algoritmo sigue la ley potencial:

$$T(n) = a \cdot n^b$$

entonces, al aplicar logaritmo a ambos lados:

$$\log T(n) = \log a + b \cdot \log n$$

Esta ecuación tiene la forma de una **línea recta** $y = m \cdot x + c$, donde:
- $y = \log T(n)$ (logaritmo del tiempo)
- $x = \log n$ (logaritmo del tamaño)
- $m = b$ es la **pendiente**, que corresponde al **exponente de la complejidad**
- $c = \log a$ es la intersección con el eje $y$

Por lo tanto, si graficamos $\log T$ vs. $\log n$ (**escala log-log**) y ajustamos una línea recta, la **pendiente de esa recta** nos dice el orden de crecimiento:
- Pendiente ≈ 1 → $\mathcal{O}(n)$
- Pendiente ≈ 2 → $\mathcal{O}(n^2)$
- Pendiente ≈ 3 → $\mathcal{O}(n^3)$

````python
# ============================================================
# Paso 6.1: Calcular los logaritmos (base 10)
# ============================================================

# TODO: Completa las siguientes líneas
# Calcula el logaritmo base 10 de cada tamaño
log_n = [math.log10(n) for n in tamanios]

# TODO: Calcula el logaritmo base 10 de cada tiempo medido
log_t = None  # Reemplaza con la lista de log10(tiempo)

print(f"{'log10(n)':>12}  {'log10(t)':>14}")
print("-" * 30)
for ln, lt in zip(log_n, log_t):
    print(f"{ln:>12.4f}  {lt:>14.4f}")
````

````python
# ============================================================
# Paso 6.2: Ajuste lineal (regresión) en escala log-log
# ============================================================

# numpy.polyfit ajusta un polinomio de grado 1 (línea recta)
# a los datos (log_n, log_t) por mínimos cuadrados.
# Retorna [pendiente, intercepto]

coeficientes = np.polyfit(log_n, log_t, 1)
pendiente    = coeficientes[0]
intercepto   = coeficientes[1]

print(f"Pendiente  (b) = {pendiente:.4f}")
print(f"Intercepto (a) = {10**intercepto:.6f}")
print(f"\nModelo ajustado: T(n) ≈ {10**intercepto:.2e} · n^{pendiente:.2f}")
````

````python
# ============================================================
# Paso 6.3: Gráfico log-log
# ============================================================

# Genera los puntos de la recta ajustada
log_n_array  = np.array(log_n)
recta_ajuste = pendiente * log_n_array + intercepto

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Gráfico 1: Escala lineal ---
axes[0].plot(tamanios, tiempos, 'o-', color='steelblue',
             linewidth=2, markersize=8, label='Tiempo medido')
axes[0].set_xlabel('Tamaño del arreglo (n)', fontsize=12)
axes[0].set_ylabel('Tiempo de ejecución (s)', fontsize=12)
axes[0].set_title('Tiempo vs. Tamaño (escala lineal)', fontsize=13)
axes[0].legend()
axes[0].grid(True, linestyle='--', alpha=0.6)

# --- Gráfico 2: Escala log-log ---
axes[1].plot(log_n, log_t, 'o', color='steelblue',
             markersize=8, label='Datos observados')
axes[1].plot(log_n_array, recta_ajuste, '--', color='tomato',
             linewidth=2, label=f'Ajuste lineal (pendiente={pendiente:.2f})')
axes[1].set_xlabel('log₁₀(n)', fontsize=12)
axes[1].set_ylabel('log₁₀(T)', fontsize=12)
axes[1].set_title('Gráfico log-log: T vs. n', fontsize=13)
axes[1].legend()
axes[1].grid(True, linestyle='--', alpha=0.6)

plt.suptitle('Análisis de Complejidad — Ordenamiento por Burbuja', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('grafico_burbuja_loglog.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Gráfico guardado como 'grafico_burbuja_loglog.png'")
````
